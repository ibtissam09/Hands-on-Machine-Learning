# Normalization data for Machine Learning

## Z-Score Normalization (Standarization)

__Why use Z-Score Normalization?__
* Centers data around the mean.
* Handles outliers better.
* Useful when features have different units or scales.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load a realistic dataset
data = pd.read_csv('../data/airline_passengers.csv')

# Display the first few rows
print(data.head())
```

```python
# Selecting the 'Passengers' column for normalization
passenger_data = data[['Passengers']]
```

```python
# Apply Z-score normalization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(passenger_data)
```

```python
# Add the normalized data back to the dataframe
data['Passengers_Zscore'] = scaled_data
```

```python
# Display the updated dataframe
print(data.head())
```

```python
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot original data on the first subplot
axs[0].plot(data['Month'], data['Passengers'], color='blue')
axs[0].set_title('Original Data')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Passengers')
axs[0].tick_params(axis='x', rotation=45)

# Plot normalized data on the second subplot
axs[1].plot(data['Month'], data['Passengers_Zscore'], color='red')
axs[1].set_title('Z-Score Normalized Data')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Z-Score')
axs[1].tick_params(axis='x', rotation=45)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

## Min-Max Normalization

__Why use Min-Max Normalization?__
* Scales all features to the same range.
* Preserves relationships within the data.
* Useful when the features are in different units or scales.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Apply Min-Max normalization
min_max_scaler = MinMaxScaler()
min_max_scaled_data = min_max_scaler.fit_transform(passenger_data)
```

```python
# Add the normalized data back to the dataframe
data['Passengers_MinMax'] = min_max_scaled_data

# Display the updated dataframe
print(data.head())
```

## Log Normalization

**Why use Log Normalization?**
* Reduces skewness in the data.
* Compresses the range of large values.
* Useful for features with exponential distributions.

```python
import numpy as np

# Apply Log normalization
data['Passengers_Log'] = np.log1p(passenger_data)
```

```python
# Display the updated dataframe
print(data.head())
```
