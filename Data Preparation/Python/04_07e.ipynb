# Recurisive Feature Elimination

### 1. Import necessary libraries

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score, precision_score
import matplotlib.pyplot as plt
```

### 2. Load the Breast Cancer dataset

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

### 3. Standardize the features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### 4. Split the dataset into training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### 5. Train a Logistic Regression model on all features (before RFE)

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)
```

### 6. Make predictions and calculate accuracy and precision


```python
y_pred = model.predict(X_test)
accuracy_before_rfe = accuracy_score(y_test, y_pred)
precision_before_rfe = precision_score(y_test, y_pred)
```

### 7. Apply RFE to reduce the number of features (selecting top 5 features)

```python
rfe = RFE(model, n_features_to_select=4)
rfe.fit(X_train, y_train)
```

### 8. Train a Logistic Regression model on the reduced feature set (after RFE)

```python

X_train_rfe = rfe.transform(X_train)
X_test_rfe = rfe.transform(X_test)
model_rfe = LogisticRegression(max_iter=10000)
model_rfe.fit(X_train_rfe, y_train)
```

### 9. Make predictions and calculate accuracy and precision after RFE

```python

y_pred_rfe = model_rfe.predict(X_test_rfe)
accuracy_after_rfe = accuracy_score(y_test, y_pred_rfe)
precision_after_rfe = precision_score(y_test, y_pred_rfe)
```

```python
accuracy_after_rfe, accuracy_before_rfe
```

### 10. Plot the results in a line chart

```python
metrics = ['Accuracy', 'Precision']
after_rfe = [accuracy_before_rfe, precision_before_rfe]
before_rfe = [accuracy_after_rfe, precision_after_rfe]

# Plotting the line chart
plt.figure(figsize=(8, 6))
plt.plot(metrics, before_rfe, marker='o', label='Before RFE', color='blue')
plt.plot(metrics, after_rfe, marker='o', label='After RFE', color='green')
plt.xlabel('Performance Metric')
plt.ylabel('Score')
plt.title('Model Performance Before and After RFE')
plt.legend()
plt.grid(True)
plt.show()
```
